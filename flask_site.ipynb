```python
#################Attempt 4############################
from distutils.log import debug
from fileinput import filename
from flask import *  
from flask_ngrok import run_with_ngrok
from flask.templating import render_template
import cv2
template_folder='/content'
app = Flask(__name__,template_folder=template_folder)  
run_with_ngrok(app)
cnn_model=load_model('/content/drive/MyDrive/FINAL_MODEL1.h5')
@app.route('/')  
@app.route('/emotion')
def main():  
    return render_template("index.html")  
  
@app.route('/success', methods = ['POST'])  
def success():  
    if request.method == 'POST':  
        f = request.files['file']
        f.save('PHOTO.png')  
        frame=cv2.imread('/content/PHOTO.png')
        #print(frame)
        #cv2_imshow(frame)
        #print(frame)
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        #cv2_imshow(gray)
        roi_gray=cv2.resize(gray,(50,50))
        #cv2_imshow(roi_gray)
        roi=roi_gray.astype('float')/255.0
        roi=img_to_array(roi)
        roi=np.expand_dims(roi,axis=0)
        #cv2_imshow(roi)
        #roi=np.expand_dims(roi,axis=0)
        pred=cnn_model.predict(roi)[0]
        print(pred)
        labels=['angry','disgusted','fearful','happy','neutral','sad','surprised']
        for x in range(7):
          for y in range(7):
            if(pred[x]<pred[y]):
              t=pred[x]
              t1=labels[x]
              pred[x]=pred[y]
              labels[x]=labels[y]
              pred[y]=t
              labels[y]=t1
        print(pred[6])
        print(labels[6]) 
        return render_template("ack.html", name = labels[6], prob=pred[6])  
  
if __name__ == '__main__':  
    app.run()
```
